```python
import numpy as np
from keras.models import Sequential
from keras.layers import Dense, ReLU,Activation
```

```python
#get the dataset and process it
import pandas as pd
import os 

#This notebooks takes dataset containing credit information and predicts credit worthiness using ANN

df = pd.read_csv("crx.txt",header= None)
df.to_csv('./data.csv',index=False)
#Check for missing data
print(df.isnull().values.any())
df

```

```python
#Get the labels
y_label = df[:][15].to_numpy()

#Convert labels to ints + = 1  - = 0 
for i,val in enumerate(y_label):
    if val == '+':
        y_label[i] = 1
    else:
        y_label[i] = 0

#Drop label columns from dataframe, its now only X
df = df.drop([15], axis =1)
df = pd.get_dummies(df)
```

```python
df
```

```python
#Dataframe to X data (labels already dropeped)
x_data = df.to_numpy()
print("Labels Shape", np.shape(y_label))
print("Samples shape " ,np.shape(x_data))
```

```python
#Architecture 1
model = Sequential(name="Credit Card NN Model 1")
model.add(Dense(32,input_dim=x_data.shape[1],activation='relu'))
model.add(Dense(16,activation='relu'))
model.add(Dense(1,activation='sigmoid'))
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
#Print model summary (Problem 1 ii)
model.summary()
model.fit(x_data,y_label,epochs=100,batch_size=100,verbose=1)
```

```python
#Architecture 2 - with Relu Activation after each layer, using Sigmoid activation at last layer same as Architecture 1
model2 = Sequential(name="Credit Card NN Model 2")
# 15 features = x_data.shape[1]
model2.add(Dense(256,input_dim=x_data.shape[1]))
model2.add(ReLU())
model2.add(Dense(128))
model2.add(ReLU())
model2.add(Dense(64))
model2.add(ReLU())
model2.add(Dense(1))
model2.add(Activation('sigmoid'))
model2.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
#Print model summary (Problem 1) ii) 
model2.summary()
```

```python
# Train with 500 epochs instead of 100 as with architecture 1 (Different hyper paramaters as 1) iii) 
model2.fit(x_data,y_label,epochs=500,batch_size=100,verbose=1)
```
